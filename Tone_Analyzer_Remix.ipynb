```python
# Keerthana's Tone Analyzer Remix
def analyze_tone():
    text = input("Enter the message you want to analyze: ").lower()
    words = text.split()

    professional = ["regards", "noted", "scheduled", "confirm", "sincerely", "update"]
    demanding = ["immediately", "urgent", "now", "must", "deadline", "asap"]
    apologetic = ["sorry", "apologize", "mistake", "pardon", "regret", "forgive"]
    assertive = ["believe", "confident", "sure", "decided", "will", "certain"]
    politeness = ["please", "thank you", "sorry"]
    uncertainty = ["maybe", "probably", "i think", "might"]
    certainty = ["definitely", "clearly", "always"]
    request = ["could you", "can you", "would you"]
    urgency = ["asap", "immediately", "now"]
    contrast = ["but", "however", "although"]

    # Scoring logic
    scores = {
        "Professional": sum(1 for w in words if w in professional),
        "Demanding": sum(1 for w in words if w in demanding),
        "Apologetic": sum(1 for w in words if w in apologetic),
        "Assertive": sum(1 for w in words if w in assertive),
        "Politeness": sum(1 for w in words if w in politeness),
        "Uncertainty": sum(1 for w in words if w in uncertainty),
        "Certainty": sum(1 for w in words if w in certainty),
        "Request": sum(1 for w in words if w in request),
        "Urgency": sum(1 for w in words if w in urgency)
    }

    # Calculate total hits to find percentages
    total_hits = sum(scores.values())

    print(f"\n--- ðŸ“Š Tone Analysis Dashboard ---")
    if total_hits == 0:
        print("Result: Neutral. No specific tone keywords detected.")
    else:
        # Sort scores to show the most dominant tone first
        sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)

        for tone, score in sorted_scores:
            percentage = (score / total_hits) * 100
            # Create a simple visual loading bar
            bar = "â–ˆ" * int(percentage / 10)
            print(f"{tone:12}: {bar} {percentage:.1f}%")

        print(f"\nPrimary Tone: {sorted_scores[0][0]}")

analyze_tone()
```
